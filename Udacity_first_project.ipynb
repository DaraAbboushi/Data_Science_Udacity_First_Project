2-What are the top 10 jobs that have been in high demand in the job market?


```python
job_type = salaries['job_title'].value_counts()[:10]

job_type
```

```python
# we are going to get the frequency for each job type to change the x-ticks and get them printed on the graph to make things clear!
n_types = salaries['job_title'].value_counts().sum()
#get the hgihest number of them 
max_job_type = job_type.shape[0]
#we get the max_prop for the xticks 
max_prop = max_job_type/n_types
#now we create an array to loop through it
tick_props = np.arange(0, max_prop+0.22, 0.01)
#make of the array, a text of each proportion in a list looped through it
tick_names = ['{:0.2f}'.format(i) for i in tick_props]
```

```python
#we will print the frquency numbers of each job title on the graph
plt.rcParams['font.size'] = 6
sb.countplot(data= salaries, y='job_title', order=job_type.index)

for i in range(job_type.shape[0]):
    count = job_type[i]
    pct_string = '{:0.2f}'.format(count*100/n_types)
    plt.text(count+1, i, pct_string, va='center')

plt.xticks(tick_props * n_types, tick_names)
plt.xlabel('Proportion')
```

Explanation: I used a bar chart to map the top 10 level jobs(categorical variable) alongside their proportion values in the dataset.

Analysis: From the bar graph, it can be seen that Data Engineer, Data scientist, Machine Learning Engineer and Data Analyst are the most demanding jobs in the Data Science job sector which is not surprising at all. It is known the data science field has been dominating the job market for years.

5- Does the job title have an effect on whether the salary is high or low for employees and how does that relate to the company size? 

```python
#i copied the the original dataset to work on it seperately without making changes to the original one.
df = salaries.copy()
#By specifying a seed value, the function makes sure that the produced sequence of random numbers remain the same for multipule runs.
np.random.seed(8000)
#i set a sample of 4000 observation to be chosen from randomly that comply with a number of jobs of my choice(i chose them randomly)
sample = np.random.choice(df.shape[0], 4000, replace= False)

# to make the subset i had to use .loc which is typically used for label indexing, loc object returns the 4000 rows and its associated columns from that DataFrame
jobs_subset = df.loc[sample]

# job titles in a list 
jobs = ['Data Engineer',
 'Data Scientist',
 'Data Analyst',
 'Machine Learning Engineer',
 'Research Engineer',
 'Applied Scientist',
 'Business Intelligence Engineer',
 'Analytics Engineer',
 'Research Scientist',
 'Principal Data Scientist',
 'Data Manager',
 'Data Architect',
 'Research Analyst',
 'Decision Scientist',
 'Machine Learning Scientist',
 'Business Intelligence Analyst',
 'Data Developer']

#let us visualise the subset
jobs_subset
```

```python
# it gives back ordered catgorical types of jobs
job_classes = pd.api.types.CategoricalDtype(ordered=True, categories=jobs)

# # I used pandas.astype() function to convert the the job_title column in the subset into an ordered categorical type
jobs_subset['job_title'] = jobs_subset['job_title'].astype(job_classes)

# I used group_by() and size() to get the number of job types and each combination of the two variable levels as a pandas Series
ct_counts = jobs_subset.groupby(['job_title', 'experience_level']).size()

# I used reset_index() to the series in the code above to convert a series into a dataframe object
ct_counts = ct_counts.reset_index(name='count')

# I used DataFrame.pivot() to rearrange the data, to have job titles on rows
ct_counts = ct_counts.pivot(index = 'job_title', columns = 'experience_level', values = 'count')

#i set the size of the white space for the figures to be set side by side
plt.figure(figsize=[23,5])
plt.subplot(1,2,1)
sb.boxplot(data = jobs_subset, x = 'job_title', y = 'salary_in_usd')
plt.xticks(rotation= 90)
plt.rcParams['font.size'] = 6

plt.subplot(1,2,2)

ct_counts = jobs_subset.groupby(['job_title', 'company_size']).size()
ct_counts = ct_counts.reset_index(name='count')
ct_counts = ct_counts.pivot(index = 'job_title', columns = 'company_size', values = 'count')
ct_counts
sb.heatmap(ct_counts, annot = True, fmt = 'd')
```

Explanation: Firstly, i chose a list of job titles contained in a list(code above) in order to map their salaries across the years indifferent to their experience level. Secondly, to avoid over plotting i set a sample of 4000 observations to be chosen from randomly but which comply with the chosen list of job titles. Thirdly, i mapped subplots side by side in order to infer to the 4th question and to explain why outliers might have occurred. Fourthly, i did not remove the outliers for that it might tell us something important and that's why i wanted to investigate it by mapping subplots side by side.

Analysis: There are many observations that one can conclude from these subplots: Firstly, the left subplot shows 50% of jobs that are enlisted in the graph, their maximum salary ranges between 280000 and 400000 usd while 50% of their average salary is above 100000. Secondly, more than a half of the jobs(in the graph) minimum is under 100000. Thirdly, there are extremely high salaries(outliers) in some jobs that are above the job's maximum salary especially in the data scientist job. Fourthly, in the right subplot, the graph depicts that all the jobs specifically Data scientist, Data Engineer, Data Analyst and Machine Learning Engineer work mostly in medium sized companies and large sized companies in comparison to small companies. The outliers found in the graph above might be explained by this heat map graph, by which the outliers for extremely high number of salaries above their maximum number work in medium companies. As a result, the job title and the company size matter has an effect in getting higher salary.
